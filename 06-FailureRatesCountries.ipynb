```python
import re
import numpy as np
import pandas as pd
import calendar
from datetime import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.cm as cmlib
import matplotlib.ticker as plticker
import datetime as dt2
import seaborn as sns
from sklearn.linear_model import LinearRegression

sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})
plt.rc('font', size=14)
plt.rc('xtick', labelsize=14) 
plt.rc('ytick', labelsize=14) 
plt.rc('axes', labelsize=14)
```

```python
df2 = pd.DataFrame([["Auckland",0.27],["US & Canada  \n (Folkman, 2018)",0.22]
                    ,["Gothenburg, Sweden \n (Malm et al., 2012)",0.16],
                   ["Small town in France \n (Jafar et al., 2010)",0.20],
                   ["Medium town in UK \n (Jara-Arriagada & \n Stoianov, 2021)",0.11],
                   ["Large city in UK \n (Rezaei et al., 2015)",0.16],
                   ["Two Australian utilities \n (Gould et al., 2011)",0.33],
                   ], columns=["Country","Failure rate"])



df2=df2.sort_values(by="Failure rate")
```

```python
fig1, ax = plt.subplots(figsize=(20,6))

colors = [ cmlib.get_cmap('cividis')(x) for x in np.linspace(0, 0.5, 9) ]
colors = colors[0:5]+[cmlib.get_cmap('cividis')(1.0)] + colors[5:10]
bars = ax.bar(df2.Country, df2['Failure rate'],color=colors)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
plt.yticks([])
#ax.set_title('Failure rate for different pipe materials')
ax.set_xlabel('Failure rate (Failures/Km/year)')
ax.xaxis.set_label_position('top') 

for bar in bars:
	ax.text(
	bar.get_x() + bar.get_width() / 2,
	bar.get_height()+0.005,
	round(bar.get_height(), 3),
	horizontalalignment='center'
)
    
fig1.savefig('Maps_Graph/FailureAnalysis/'+'FailureRateOtherCountries' +'.png', dpi=200, bbox_inches='tight')

```
